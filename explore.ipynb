NaT = Not a Time — pandas' equivalent of NULL for datetime/timedelta columns.

For PitInTime and PitOutTime it means: this lap had no pit stop.

A stint is the period of continuous driving on one set of tires between pit stops (or from race start to first pit, or last pit to finish).

Race start

├── Stint 1: Laps 1-18  (Soft tires) ↓ PIT STOP LAP 18

├── Stint 2: Laps 19-38 (Medium tires)  ↓ PIT STOP LAP 38

└── Stint 3: Laps 39-57 (Hard tires)
         ↓ Finish

PitInTime lives on the last lap of a stint — the lap you drove INTO the pits.

PitOutTime lives on the first lap of the next stint — the lap you drove OUT of the pits.

Logic for Calculation of PitStop Time, we get these times, then we convert it to seconds for Snowflake

If it starts raining at lap 32 but your 90 second bucket starts at lap 31.5. That lap gets Rainfall=False even though it rained mid-lap. For a dry Bahrain race that doesn't matter. For a wet Silverstone race that completely changes the strategy story.

Reducing granularity to match your primary dataset before joining is cleaner than an asof join and easier to replicate in dbt SQL later. The rainfall edge case is the only real gotcha and max() within the bucket actually handles it correctly — if it rained at any point in that 90 second window, the lap gets Rainfall=True.

Aggregating to 90 second buckets assumes all laps are exactly 90 seconds and perfectly aligned to your buckets. In reality:

Lap 1 might be 105 seconds (formation, slower)
Safety car laps can be 120+ seconds
Pit stop laps are longer
Buckets won't align perfectly with when each lap actually started

So you could end up matching a lap to weather from 45 seconds before or after the lap actually happened.
AsOf join finds the closest actual weather reading to that exact lap timestamp — no bucket alignment problem, no averaging away real variation.

For TrackTemp and AirTemp — probably not much. Temperature changes slowly. A 30-45 second mismatch gives you ~0.1°C error. Negligible.
For Rainfall — it matters a lot. Rain can start and stop in minutes. AsOf is safer here.

I initially considered bucketing weather to 90 second intervals but realised lap lengths vary significantly — safety car laps, pit laps, lap 1 — so bucket alignment would introduce timing errors. AsOf join gives each lap the closest actual weather reading regardless of lap length.

GridPosition — where the driver started the race. Their qualifying result translated to a grid slot. Verstappen on pole = GridPosition 1.

Position — where the driver finished the race on the road. Pure physical crossing order at the chequered flag.

ClassifiedPosition — the official FIA classified result after stewards have applied rules. This is what actually goes in the championship standings.

Use ClassifiedPosition for championship points calculations — it's the official result

Use Position for race narrative analysis — it shows what actually happened on track

Use GridPosition for "positions gained" metric — GridPosition - ClassifiedPosition tells you how many places a driver gained or lost during the race

DNFs — a driver who retires on lap 23 of a 57 lap race only has 23 lap rows. Their session ends there.

Lap 1 incidents — a driver who crashes on the formation lap or lap 1 might have 0 or 1 laps.

Classified but retired — some drivers are classified finishers despite not completing the full race distance, so their lap count is lower than the race winner's.

Lapped drivers — interestingly this goes the other way. A driver who gets lapped actually completes the same number of laps as the leader because the race ends when the leader crosses the line — everyone stops at that same lap count. So lapping doesn't cause unequal lap counts.
